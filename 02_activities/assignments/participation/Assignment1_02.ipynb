```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Line plot with both y variables on same axes

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x,y1,y2)
```

```python
# add legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc= 'lower right') #loc - location
```

```python
# Modify our legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc= 'lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, # specify no of column
          shadow = True)
```

```python
# Move legend outside the plot

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc= 'upper left', # refer to the corner of the legend
          bbox_to_anchor = (1,1)) #specify where legend should be using coordinates - here move outside the axis
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc= 'upper left', # refer to the corner of the legend as anchor
          bbox_to_anchor = (1,0.5)) # 50 % on y axis
```

```python
# Add Text Annotation

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc= 'lower right')

#ax.text(10, 95, "This value is important!") #10 = horizontal poisition and 96 = vertical position
# position for starting text annotation

# Modifying text annotations (appearance)

ax.text(10,95, "This value is important!",
        ha = 'center', #horizontal algnment of text
        color = 'red', #font color
        size = 20) #font size
```

```python
# position text annotations using axes object but not related to data

fig, ax =plt.subplots()
ax.axis([0,10,0,10]) # define axis range for x and y

ax.text(1,5, "Data:(1,5)", transform=ax.transData) #according to data
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes) # according to axis
ax.text(0.2, 0.2, ". figure:(0.2, 0.2)",transform=fig.transFigure) 
# according to the figure box 20% up and 20% down
```

```python
# Annotating with arrows

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc= 'lower right')

ax.annotate('This is important!',
            xy=(10,95), # xy location of where arrow should go
            xytext=(20,94), # location of where text should go
            arrowprops=dict(facecolor='black',shrink=0.05) #how arrow looks like
            ) # shrink size, distance between text and arrow
```

```python
# Modifying arrowstyle

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc= 'lower right')

ax.annotate('This is important!',
            xy=(10,95), 
            xytext=(20,94),
            arrowprops=dict(arrowstyle = "wedge",
                            color = 'hotpink'))
```

```python
# Remove tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # remove ticks and labels for y axis
ax.xaxis.set_major_formatter(plt.NullFormatter()) # remove labels only for x axis
```

```python
# Limit tick marks

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # specify the number of tick marks and label for x axis
```

```python
# Set tick mark interval

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # set tick marks at a particular interval - 5 
```

```python
# Rotating axis labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right') # rotate labels by 45 degree ub rught alignment to give more space

# Motify axis title font

# font dictionary
font1 = {'family':'serif','color':'indigo'}
# assign xlabel
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict=font1)
```
