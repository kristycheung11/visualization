```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Subplots

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))

ax1.scatter(x1,y1,color='indigo')
ax2.bar(x2,y2)

# Activity - Customize plot
ax1.set_title('Scatter Plot', loc = 'center') 
ax2.set_title('Bar Plot', loc = 'center') 

ax1.xaxis.set_major_locator(plt.MultipleLocator(10))
```

```python
# Define subplots using subplot mosaic

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], #someaxes is a key term for using mosaic
                                    ['ax2', 'ax3']], figsize=(7, 4), # this define the positioning of the plot
                                    layout= "constrained") # add layout type to make sure everything fits

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
# Multiple viz on one axes

# sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define figure and axes
fig, ax = plt.subplots(figsize=(7,3))

# add bar and line plot to the same axes
ax.bar(x, y1, color="indigo")
ax.plot(x, y2, color ="red")
```

```python
# add error information

# calculate standard deviation of data
y2_sd = np.std(y2)

#plot
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color= "red")

#add error info
ax.errorbar(x, # x-value
            y2, # y-value
            yerr = y2_sd, # specify we want vertical error bars
            fmt ="none") # format none means we do not want the data points just the bar
```

```python
# Customize errorbar appearance

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color= "red")

ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = "indigo", # error bar color
            elinewidth = 4, # thickness of errorbar
            capsize = 6, # how horizontally wide are the caps of the bars
            capthick = 4,  # how vertically wide are the caps
            errorevery = 2) # specify intervals for error bar
```

```python
# Add images to plot

#load library
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images

```

```python
# Define image variables

response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png') 
# use request to get the image from the specific url and store in variable response
image_file = BytesIO(response.content) # store and save the url image in a file
image = Image.open(image_file) # open and render image
```

```python
# basic plot
fig, ax = plt.subplots (figsize=(7,3))
ax.plot (x, y2, color ='red')

# overlay a new axis ax_image as container for image
ax_image =fig.add_axes([0.1, # define x coordinate on figure
                        0.11, # define y coordinate on figure
                        0.15, # image width in percentage of original
                        0.35]) # image height

# add image to the container
ax_image.imshow(image) # show image
ax_image.axis('off') # get rid of the border for the image
```
